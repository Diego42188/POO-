```python
#
import matplotlib.pyplot as plt

class Punto():
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def plot(self):
        plt.scatter(self.x, self.y, color='red', marker='o')
        plt.xlabel('X-axis')
        plt.ylabel('Y-axis')
        plt.title('Gr√°fica del punto')
        plt.grid(True)
        plt.show()
```

```python
punto = Punto(2, 8)#acabamos de construir al punto (2, 8)
```

```python
punto.plot()
```
